```python
#Mini Project 2-Polynomial Regression

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Given data
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 15, 27, 40]

# Convert to numpy array and reshape for sklearn
x = np.array(x).reshape(-1, 1)
y = np.array(y)

# Choose the degree of the polynomial (try degree=2, 3, or higher based on data pattern)
degree = 3
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)

# Fit the model
model = LinearRegression()
model.fit(x_poly, y)

# Predict for visualization
x_range = np.linspace(min(x), max(x), 100).reshape(-1, 1)
x_range_poly = poly_features.transform(x_range)
y_pred = model.predict(x_range_poly)

# Plot
plt.scatter(x, y, color='red', label='Original Data')
plt.plot(x_range, y_pred, color='blue', label=f'Polynomial Degree {degree}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```
